# 경사 하강법

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100)
y = X * 0.7

plt.figure(figsize=(8, 6))
plt.scatter(X, y)
```

```python
w = np.random.uniform(-1, 1)
w
```

```python
b = np.random.uniform(-1, 1)
b
```

```python
lr = 0.5

for epoch in range(300):
    y_pred = w * X + b  # 딥러닝에서는 해당 수식이 모델이라 생각하면 된다.

    error = np.abs(y_pred - y).mean()   # 손실 함수

    w_grad = lr * ((y_pred - y) * X).mean()     # 최적화 
    b_grad = lr * (y_pred - y).mean()

    w = w - w_grad      # 새롭게 가중치를 업데이트 하는 것(역전파: back propagation)
    b = b - b_grad

    print('error: ', error)

    if epoch % 10 == 0:
        y_pred = w * X + b
        plt.figure(figsize=(8, 6))
        plt.scatter(X, y)
        plt.scatter(X, y_pred)
        plt.show()
        
    if error < 0.001:
        break
```
